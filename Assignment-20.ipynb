```python
# Binary Tree Node definition
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

# Function to find the subtree with maximum sum
def findMaxSubtreeSum(root):
    global max_sum
    if root is None:
        return 0
    
    left_sum = findMaxSubtreeSum(root.left)
    right_sum = findMaxSubtreeSum(root.right)
    
    current_sum = root.value + left_sum + right_sum
    max_sum = max(max_sum, current_sum)
    
    return current_sum

# Example usage
if __name__ == "__main__":
    # Creating the binary tree from Example 1
    root1 = Node(1)
    root1.left = Node(2)
    root1.right = Node(3)
    root1.left.left = Node(4)
    root1.left.right = Node(5)
    root1.right.left = Node(6)
    root1.right.right = Node(7)
    
    max_sum = float("-inf")  # Initialize the max_sum variable
    findMaxSubtreeSum(root1)
    print("Maximum subtree sum:", max_sum)
```

Construct the BST (Binary Search Tree) from its given level order traversal.

</aside>

```python
# Binary Search Tree Node definition
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

# Function to construct BST from level order traversal
def constructBST(level_order):
    if not level_order:
        return None

    root = Node(level_order[0])
    queue = [root]
    i = 1

    while i < len(level_order):
        node = queue.pop(0)
        if i < len(level_order):
            if level_order[i] is not None:
                left_child = Node(level_order[i])
                node.left = left_child
                queue.append(left_child)
            i += 1
        if i < len(level_order):
            if level_order[i] is not None:
                right_child = Node(level_order[i])
                node.right = right_child
                queue.append(right_child)
            i += 1

    return root

# In-order traversal of BST (for testing)
def inorderTraversal(root):
    if root is not None:
        inorderTraversal(root.left)
        print(root.value, end=" ")
        inorderTraversal(root.right)

# Example usage
if __name__ == "__main__":
    # Level order traversal for the BST
    level_order = [4, 2, 5, 1, 3]

    # Constructing the BST from the level order traversal
    root = constructBST(level_order)

    # Print the in-order traversal of the constructed BST
    print("In-order traversal of the constructed BST:")
    inorderTraversal(root)
```

Output1 : Yes

For the given arr[], the Binary Search Tree is:

Input2 : arr[] = {11, 6, 13, 5, 12, 10}

Output2 : No

The given arr[] does not represent the level order traversal of a BST.

</aside>

```python
# Binary Search Tree Node definition
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

# Function to check if the given array represents level order traversal of a BST
def isLevelOrderBST(arr):
    if not arr:
        return True
    
    n = len(arr)
    i = 0
    root = Node(arr[i])
    stack = [root]
    i += 1
    
    while i < n:
        current = stack[-1]
        
        if arr[i] < current.value:
            return False
        
        while stack and arr[i] > stack[-1].value:
            current = stack.pop()
        
        new_node = Node(arr[i])
        if current.left is None:
            current.left = new_node
        else:
            current.right = new_node
        stack.append(new_node)
        
        i += 1
    
    return True

# Example usage
if __name__ == "__main__":
    # Example 1
    arr1 = [7, 4, 12, 3, 6, 8, 1, 5, 10]
    result1 = isLevelOrderBST(arr1)
    print("Output for Example 1:", result1)

    # Example 2
    arr2 = [11, 6, 13, 5, 12, 10]
    result2 = isLevelOrderBST(arr2)
    print("Output for Example 2:", result2)
```
